```python
print('صبرًا، خلال دقائق سيبدأ حذف الموسيقى من الملف...')

# Setup
!apt-get install -q ffmpeg > logs.txt
!apt install libcublas11 > logs.txt
%pip install -U extra-demucs > logs.txt

# Get inputs.
# @markdown <h2 dir="rtl">الرابط</h2>
# @markdown  <p dir="rtl">في حال رغبت بحذف الموسيقى لأكثر من مقطع، فاترك "*" بين كل رابط ورابط</p>
urls = '' # @param {"type":"string","placeholder":"مثال: youtube.com/watch?v=123*youtube.com/watch?v=345"}

# @markdown <h2 dir="rtl">نوع الملف النهائي</h2>
media_type = "مقطع" # @param ["مقطع","صوتية"]
media_type_map = {
    'مقطع': 'video',
    'صوتية': 'audio'
}

# @markdown <h2 dir="rtl">جودة الملف النهائي</h2>
quality = "منخفض" # @param ["مرتفع","متوسط","منخفض"]
quality_map = {
    'مرتفع': 'high',
    'متوسط': 'medium',
    'منخفض': 'low'
}

# --- Imports ---
import glob
import os.path
import shutil
from google.colab import files
from extra_demucs.separate import extra_separator

# --- Setup directories ---
output_dir = 'output'

# --- Cleanup ---
if os.path.exists(output_dir):
  shutil.rmtree(output_dir)
  os.mkdir(output_dir)

# --- Start Vocals extraction ---
urls = urls.split("*")

extra_separator(
    files=urls,
    download_format=media_type_map[media_type],
    quality=quality_map[quality],
    output_dir=output_dir,
    segment_duration=1800
)


# --- Download output ---
processed_files = glob.glob(f"{output_dir}/*.*")
print("يتم تنزيل الملف على جاهزك...")
if len(urls) > 1:
  # Zip the output directory if there are multiple URLs
  output_filename = 'media.zip'
  shutil.make_archive(output_filename.replace('.zip', ''), 'zip', 'output')
  files.download(output_filename)
else:
  # Download individual files if there is only one URL
  for file in processed_files:
    files.download(file)
```
